```kotlin
val correlationTable = df
    .select { Survived  and SibSp and Parch and Age  and Fare and Sex and Embarked}
    .convert{ valueCols() }.toDouble()
    .corr{ valueCols()}.withItself()
    .gather{ allAfter("column") }.into("row", "value")

correlationTable
```

```kotlin
fun scaleContinuousColorGradientN() = continuousColorGradientN(
    gradientColors = listOf(
        JetBrainsColors.orange,
        JetBrainsColors.lightGrey,
        JetBrainsColors.darkGrey,
    ),
    domainMin = -1.0,
    domainMax = 1.0,
)
```

```kotlin
correlationTable.plot {
    tiles {
        x(row) { axis.name = "" }
        y(column) { axis.name = "" }
        fillColor(value) { scale = scaleContinuousColorGradientN() }
    }
}
```

```kotlin
correlationTable.plot {
    points {
        size(value){
            legend {
                breaks(emptyList())
            }
        }
        x(row) {
            axis.name = ""
        }
        y(column) {
            axis.name = ""
        }
        color(value) { scale = scaleContinuousColorGradientN() }
    }
    layout {
        style {
            panel.grid {
                majorLine {
                    blank = true
                }
            }
        }
    }
}
```

```kotlin
%useLatestDescriptors
%use dataframe, kandy
```

```kotlin
val df = DataFrame.readCSV(
    fileOrUrl = "./data/titanic.csv",
    delimiter = ';'
)

df
```

```kotlin
df.describe()
```

```kotlin
val tempVar = df.dropNulls{age}
```

```kotlin
val ageMean =  tempVar.filter{ age < 100}.select{age}.distinct().mean()
ageMean.age
```

```kotlin
val tempVar = df.fillNulls{age}.with{ 100.0 }
```

```kotlin
val df1 = tempVar.convert{age}.with{ if (it >= 100.0) ageMean.age.toInt() else it.toInt()}
df1
```

```kotlin
df1.describe()
```

```kotlin
df1.valueCounts{sex}
```

```kotlin
df1.sibsp.distinct()
```

```kotlin
df1.parch.distinct()
```

```kotlin
df1.fare.distinct()
```

```kotlin
df1.valueCounts{age}
```

```kotlin
val df2 = df1.fillNulls { sibsp and parch and fare }.perCol{ mean() }
    .fillNulls {sex}. with{ "female" }
    .fillNulls { embarked}.with{"S"}
    .convert {sibsp and parch  and fare}.toInt()
    .convert { sex }.with{ if (it == "male") 1 else if (it =="female") 2 else 3}

df2
```

```kotlin
df2.describe()
```

```kotlin
df2.corr()
```

```kotlin
val df3 = df2
    .corr { all() }.with { survived }
    .sortBy { survived }

df3
```

```kotlin
df3.plot{
    bars {
        x(column)
        y(survived) { scale = continuous(-1.0..1.0) }
        fillColor(survived) {
            scale = continuousColorGradientN(listOf(Color.hex("#E0F7FA"), Color.hex("#006064")))
            legend { type = LegendType.None }
        }
    }
}
```

```kotlin
df2.groupBy{pclass}.mean{ survived}
    .plot{
        bars {
            x(pclass)
            y(survived){ scale = continuous(0..1) }
            fillColor(survived) {
                scale = continuousColorGradientN(listOf(Color.hex("#E0F7FA"), Color.hex("#006064")))
                legend { type = LegendType.None }
            }
        }
    }
```

```kotlin
df2.groupBy{sex}.mean{ survived}
    .plot{
        bars {
            x(sex)
            y(survived){ scale = continuous(0..1) }
            fillColor(survived) {
                scale = continuousColorGradientN(listOf(Color.hex("#E0F7FA"), Color.hex("#006064")))
                legend { type = LegendType.None }
            }
        }
    }
```

```kotlin
// JetBrains color palette
object JetBrainsColors {
    val lightOrange = Color.hex("#ffb59e")
    val orange = Color.hex("#ff6632")
    val lightGrey = Color.hex("#a6a6a6")
    val darkGrey = Color.hex("#4c4c4c")
}
```

```kotlin
val byAge = df2.valueCounts { age }.sortBy { age }
byAge
```

```kotlin
byAge.plot {
    points {
        x(age)
        y(count)
        size = 5.0
        color = JetBrainsColors.lightOrange
    }
    layout {
        size = 850 to 500
    }
}
```

```kotlin
df2.select{age}.plot {
    histogram(x = age, binsOption = BinsOption.byWidth(5.0)) {
        fillColor = JetBrainsColors.orange
    }
    layout {
        size = 850 to 500
    }
}
```

```kotlin
df2.groupBy { age }.pivotCounts { survived }.sortBy { age }
```

```kotlin
val survivedByAge = df2.select { survived and age}.sortBy{age}.convert{survived}.with{if (it == 1) "Survived" else "Died"}
```

```kotlin
survivedByAge.groupBy { survived }.plot {
    histogram(x = age, binsOption = BinsOption.byWidth(5.0)) {
        fillColor(key.survived) {
            scale = categorical(
                "Died" to JetBrainsColors.darkGrey,
                "Survived" to JetBrainsColors.orange,
            )
        }
        alpha = 0.7
        position = Position.dodge()
    }
    layout {
        size = 850 to 500
    }
}
```

```kotlin
survivedByAge.groupBy { survived }.plot {
    densityPlot(x = age) {
        fillColor = Color.GREY
        alpha = 0.3
        borderLine {
            color(key.survived) {
                scale = categorical(
                    "Survived" to JetBrainsColors.orange,
                    "Died" to JetBrainsColors.darkGrey,
                )
            }
        }
    }
    layout {
        size = 850 to 250
    }
}
```

```kotlin
survivedByAge.plot {
    boxplot(x = survived, y = age) {
        boxes {
            fillColor(Stat.x) {
                scale = categorical(
                    "Survived" to JetBrainsColors.orange,
                    "Died" to JetBrainsColors.darkGrey,
                )
            }
        }
    }
    layout {
        size = 500 to 400
    }
}
```

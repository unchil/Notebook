## Update

```kotlin
df.update{city}.at(2).with{ "St. Petersburg"}
```

```kotlin
df.update{ colsOf<String?>() }.where{age > 20}.with{ it?.uppercase()}
```

```kotlin
df.update{ colsOf<String?>()}.with{ it?.lowercase()}
```

```kotlin
df.fillNulls{ colsOf<String?>()}.with { "__" }
```

## Convert

```kotlin
df.convert{name}.with{ name.firstName + " " + name.lastName}
```

```kotlin
val convDf = df.convert { age }.toDouble()
convDf.describe()
```

```kotlin
val convDf = df.convert { age }.toStr().parse{ age}
convDf
```

```kotlin
convDf.describe()
```

## Split / Merge

```kotlin
df.split{ city }.by {it.toCharArray().toList()}.into{"c$it"}
```

```kotlin
df.split{ city }.by {it.toCharArray().toList()}.inward{"c$it"}
```

```kotlin
df.split { city }.by { it.toCharArray().toList() }.inplace()
```

```kotlin
df.split { city }.by { it.toCharArray().toList() }.intoRows()
```

```kotlin
df.convert { city }.with { it?.toCharArray()?.toList() }.explode()
```

```kotlin
df.merge{ name.firstName and name.lastName}.by(" ").into("full name")
```

```kotlin
df.merge{ name.lastName and city}.by{ it[0] + " from " + it[1]}.into("info")
```

```kotlin
df.update { age }.where { city == "Paris" }.with { it - 5 }
    .filter { isHappy && age > 100 }
    .move { name.firstName and name.lastName }.after { isHappy }
    .merge { age and weight }.by { "Age: ${it[0]}, weight: ${it[1]}" }.into("info")
    .rename { isHappy }.into("isOK")
```
